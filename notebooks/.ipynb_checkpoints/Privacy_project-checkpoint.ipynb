```python
import sys
import os

# Ensure our src package is loadable
sys.path.append(os.path.abspath('../src'))

import pandas as pd
import flwr as fl
from flwr.common import Context

from task import load_data, Net, FlowerClient, get_device

def client_fn(context: Context) -> fl.client.Client:
    """Function to create a client instance on demand."""
    # Get partition id (simulating different clients)
    partition_id = int(context.node_config.get("partition-id", 0))
    
    trainloader, valloader = load_data(partition_id=partition_id, batch_size=32)
    device = get_device()
    net = Net().to(device)
    
    return FlowerClient(net, trainloader, valloader, device).to_client()
```

```python
from flwr.server import ServerApp, ServerConfig
from flwr.server.strategy import FedAvg
from flwr.client import ClientApp

strategy = FedAvg()

def server_fn(context: Context) -> ServerApp:
    config = ServerConfig(num_rounds=3)
    return ServerApp(config=config, strategy=strategy)

client_app = ClientApp(client_fn=client_fn)
server_app = ServerApp(server_fn=server_fn)
```

```python
import flwr

print(f"\n🚀 Starting Federated Learning Simulation with Differential Privacy on: {get_device()} \n")

# Run simulation
flwr.simulation.run_simulation(
    server_app=server_app,
    client_app=client_app,
    num_supernodes=10,
    backend_name="ray",
    backend_config={"client_resources": {"num_cpus": 1, "num_gpus": 0.0}}
)
```
